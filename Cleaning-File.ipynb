```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
#removing unnecessary columns
main_df = main_df.drop(main_df.loc[:, 'Date1':' diff_pressure'].columns, axis=1)
```

```python
#remove the character " ' " from the dataframe
main_df = main_df.applymap(lambda x: x.replace("'", ''))
#main_df = main_df.replace(to_replace="'", value='')
print(main_df)
```

```python
#check and fix unclosed parenthesis in the column headers
for col in main_df.columns:
    if col.count('(') != col.count(')'):
        main_df = main_df.rename(columns={col : col+ ')'})
main_df.columns
#main_df.columns = [main_df.rename(columns={col : col+ ')'}) if col.count('(') != col.count(')') else col for col in main_df.columns]
```

```python
#slicing year, month, date into new columns
new_column = main_df['Date'].str[0:4]
new_column2 = main_df['Date'].str[5:7]
new_column3 = main_df['Date'].str[8:]
main_df['Year'] = new_column
main_df['Month'] = new_column2
main_df['Day'] = new_column3
main_df
```

```python
subset = main_df.loc[:,['Day','Month','Year']]
increment = 2022
for i in range(1, len(subset)):
    if subset.loc[i,'Month'] == '01':
        if subset.loc[i-1,'Month'] == '12' or subset.loc[i-1,'Month'] == '11':
            increment += 1
    subset.loc[i,'Year'] = increment
subset
```

```python
#search for non leap years
unique_years = subset['Year'].unique()

# Function to check if a year is a leap year or not
def NotLeapYear(input_year):
    if ((input_year % 400 == 0) or ((input_year % 4 == 0) and (input_year % 100 != 0))):
        return None
    else:
        return input_year

# List to store non-leap years
nonleap_years = []

# Iterate over the unique years
for i in unique_years:
    year = int(i)
    result = NotLeapYear(year)
    if result is not None:
        nonleap_years.append(result)

# Filter the DataFrame to include only rows where the day is 29, the month is 2, and the year is in nonleap_years
filtered_df = subset[(subset['Day'] == '29') & (subset['Month'] == '02') & (subset['Year'].isin(nonleap_years))]

# Replace the 29 values with 28 in the original DataFrame
subset.loc[filtered_df.index, 'Day'] = 28


# Verify that the replacement was successful
''' 
verification_df = subset.loc[(subset['Day'] == 29) & (subset['Month'] == 2) & (subset['Year'].isin(nonleap_years))]
if verification_df.empty:
    print("All 29 values in the 'Day' column for 2 in the 'Month' column in a non-leap year have been replaced with 28.")
else:
    print("Some or all 29 values in the 'Day' column for 2 in the 'Month' column in a non-leap year have not been replaced with 28.")'''


subset['Day'] = subset['Day'].astype(str)
subset['Month'] = subset['Month'].astype(str)
subset['Year'] = subset['Year'].astype(str)
#Merge Day, month, Year
subset['Date'] = subset['Year'] + '-' + subset['Month'] + '-' + subset['Day']
#changing the format to date
'''for i in subset['Date']:
    if i == '2029-02-29':
        print('True')'''

subset['Date'] =pd.to_datetime(subset['Date'])

main_df = main_df.drop(main_df.loc[:, 'Year':'Day'].columns, axis=1)
```
